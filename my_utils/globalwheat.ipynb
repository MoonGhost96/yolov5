```python
import os
import shutil
```

```python
dir_list = ['arvalis_1', 'arvalis_2', 'arvalis_3', 'ethz_1', 'rres_1', 'inrae_1', 'usask_1']
file_dir = '/home/hjc/datasets/GlobalWheat2020'
val_image_dir = '/home/hjc/datasets/GlobalWheat2020/images/val'
val_label_dir = '/home/hjc/datasets/GlobalWheat2020/labels/val'
```

```python
k = -1
l=0
for d in dir_list:
    path = os.path.join(file_dir,"images",d)
    images = os.listdir(path)
    for image in images:
        k+=1
        if k % 10 != 0:
            continue
        filename1 = os.path.splitext(image)[1]  # 读取文件后缀名
        filename0 = os.path.splitext(image)[0]  #读取文件名
        image_file = os.path.join(path,image)
        label_file = os.path.join(file_dir,'labels',d,filename0+'.txt')
        if os.path.exists(label_file):
            shutil.move(image_file,val_image_dir )
            shutil.move(label_file, val_label_dir)
```
